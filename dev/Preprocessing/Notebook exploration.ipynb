````python

def reformat(col):
    col2 = []
    for i in col : 
        i = ast.literal_eval(i)
        col2.append(i)
    return  reduce(operator.concat, col2)
ingrédient = reformat(echant['NER'])
ingrédient = reformat(echant['NER'])
print(f'Longueur de la liste des ingrédients: {len(ingrédient)} VS nombre d\'ingrédients: {len(set(ingrédient))}')
print(f'Les 20 ingrédients les plus utilisés dans les recettes sont : ')
Counter(ingrédient).most_common(20)
````

## savoir si plusieurs ingrédients sont dans une recette

````python
# avec la méthode .apply
test = ['olive','tomato']
echant[echant['NER'].apply(lambda sentence: any(word in sentence for word in test))] # renvoie les recettes contenant un des deux ingrédients
echant[echant['NER'].apply(lambda sentence: all(word in sentence for word in test))] # les deux
````

````python
# avec la méthode reduce
echant[reduce(lambda a, b: a&b, (echant['NER'].str.contains(s) for s in test))]
````

````python
# avec une expression régulière
base = r'^{}'
expr = '(?=.*{})'
echant[echant['NER'].str.contains(base.format(''.join(expr.format(w) for w in test)))][['title','NER']]
````

## Avoir le pourcentage d'ingrédients déjà à dispositon

````python
nb = len(test)
rep = echant[echant['NER'].str.contains(base.format(''.join(expr.format(w) for w in test)))][['title','NER']]
rep['%'] = rep['NER'].apply(lambda sentence: round((nb / len(ast.literal_eval(sentence)))*100,1))
rep.sort_values('%',ascending=False)
````
